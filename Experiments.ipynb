## Compute PC-ER  

```python
# Ablation emb


import pandas as pd
import numpy as np
import warnings

# Filter out the specific warning
warnings.filterwarnings("ignore", message="invalid value encountered in divide")
warnings.filterwarnings("ignore", category=RuntimeWarning)

datasets = ['assist09_tkde', 'assist17_tkde', 'algebra','math_1', 'math_2']
methods = ["BPR","DINA","NCDM","MIRT","MCD"]

for dataset in datasets :
    print(f'--{dataset}--')
    for method in methods:
        print(f'{method} :')
        embs = []
        d = []
        for i_split in range(5):
            
            embs.append(pd.read_csv(
                "./results/table_2/users/"+dataset+"_" + str(
                    i_split) + "_"+str(method)+".csv", header=None).to_numpy())
            d.append(pd.read_csv(
                "./data/cdbpr_format/"+dataset+"/train_valid_" + str(i_split) + ".csv",
                header=None, sep=",").to_numpy())
        embs_array = np.array(embs)
        r = np.zeros(shape=(5, embs_array.shape[1]))
        
        for i_split in range(5):
            b_reshaped = embs_array[i_split, :, :]
        
            U_resp_sum = np.zeros(shape=(embs_array.shape[1], embs_array.shape[2]))  # nb_user,nb_dim
            U_resp_nb = np.zeros(shape=(embs_array.shape[1], embs_array.shape[2]))
            for l in d[i_split]:
                U_resp_sum[l[0], l[3]] += l[2]
                U_resp_nb[l[0], l[3]] += 1
            ave_r = (U_resp_sum / U_resp_nb)
        
            for i_user, u in enumerate(ave_r):
                u_clean = u[~np.isnan(u)]
                b_clean = b_reshaped[i_user, :][~np.isnan(u)]
                u_clean = u_clean[~np.isnan(b_clean)]
                b_clean = b_clean[~np.isnan(b_clean)]
                r[i_split][i_user] = np.corrcoef(u_clean, b_clean)[0][1]
        print(f'{np.nanmean(r,axis=1).mean()}+-{np.nanmean(r,axis=1).std()}')
```
